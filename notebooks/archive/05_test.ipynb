```python
import pandas as pd
import numpy as np

transactions = pd.read_pickle(r'data\03_int\int_transactions.pkl')
offers = pd.read_pickle(r'data/03_int/int_transcript_offers.pkl')
```

```python
#Filter on completed offers
completed_offers = offers[offers['offer_completed'].notna()]
completed_offers.head()
```

```python
def get_customer_data(customer_id, df, output_path):
    output_df = df[df['customer_id'] == customer_id]
    output_df.to_csv(output_path, index=False)
    return output_df

co_customer1 = get_customer_data(
                                '0009655768c64bdeb2e877511632db8f', 
                                completed_offers, 
                                r'data\03_int\int_example_customer_1_completed_offers.csv'
                                )
co_customer1.head()
```

```python
trans_customer1 = get_customer_data(
                                '0009655768c64bdeb2e877511632db8f', 
                                transactions, 
                                r'data\03_int\int_example_customer_1_transactions.csv'
                                )
trans_customer1.head()
```
