```python
import pandas as pd
```

```python
transcript = pd.read_pickle(r'data\02_stg\stg_transcript.pkl')
portfolio = pd.read_pickle(r'data\02_stg\stg_portfolio.pkl')
transcript.head()
                            
```

```python
#Split transcript into 2 dataframes: offer and transaction
offers = transcript[transcript['event'].isin(['offer received', 'offer viewed', 'offer completed'])]
offers = offers.drop(columns=['transaction_amount','reward'])

#Make time_hrs the first column
cols = list(offers.columns)
cols = [cols[-1]] + cols[:-1]
offers = offers[cols]

#Left join portfoliop['offer type'] to transcript on 'offer_id'
offers = offers.merge(portfolio[['offer_id', 'offer_type', 'duration_hrs', 'difficulty']], on='offer_id', how='left')

offers.head()
```

```python
example_customer_offers = offers[offers['customer_id'] == '0009655768c64bdeb2e877511632db8f']

example_customer_offers.to_csv(r'data\03_int\int_example_customer_offers.csv', index=False)
example_customer_offers.head()
```

```python
transactions = transcript[transcript['event'] == 'transaction']
transactions = transactions.drop(columns=['offer_id','reward', 'event'])
transactions.reset_index(inplace=True)

#Reorder columns
cols = ['time_hrs', 'customer_id', 'transaction_amount']
transactions = transactions[cols]

transactions.head()
```

```python
example_customer_transactions = transactions[transactions['customer_id'] == '0009655768c64bdeb2e877511632db8f']

example_customer_transactions.to_csv(r'data\03_int\int_example_customer_transactions.csv', index=False)
example_customer_transactions.head()
```
