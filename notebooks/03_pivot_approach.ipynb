```python
transcript = pd.read_pickle(r'data\02_stg\stg_transcript.pkl')

#Create a new column to store the event_id
transcript = transcript.reset_index()
transcript = transcript.rename(columns={'index': 'event_id'})
transcript.head()
```

```python
# One-hot encode the 'event' column
transcript_encoded = pd.get_dummies(transcript, columns=['event'])

# Define a dictionary mapping old column names to new ones
rename_dict = {
    'event_offer received': 'offer_received',
    'event_offer viewed': 'offer_viewed',
    'event_offer completed': 'offer_completed',
    'event_transaction': 'transaction'
}

# Rename the columns
transcript_encoded = transcript_encoded.rename(columns=rename_dict)

# #Reorder columns
cols = [
    'event_id', 'time_hrs', 'customer_id', 'offer_id',
    'offer_received', 'offer_viewed', 'offer_completed', 'transaction',
    'transaction_amount', 'reward']

transcript_encoded = transcript_encoded[cols]

# Display the first few rows
transcript_encoded.head()
```

```python
example_customer_transcript_encoded = transcript_encoded[transcript_encoded['customer_id'] == '0009655768c64bdeb2e877511632db8f']

example_customer_transcript_encoded.to_csv(r'data\03_int\int_example_customer_transcript_encoded.csv', index=False)
example_customer_transcript_encoded.head()

```
